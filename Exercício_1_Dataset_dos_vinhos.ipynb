```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#Carregando a base de dados dos vinhos
wine = datasets.load_wine()
#Definindo o X que são as informações que temos para predizer o Y
x = wine.data
#Definindo o Y que é o resultado que queremos descobrir
y = wine.target
```

```python
#Printando os valores de X
x
```

```python
#Printando os valores de Y
y
```

```python
# Criando a primeira árvore de decisão do dataset de vinhos
clf_wine = DecisionTreeClassifier(random_state=1234)
#Treinando a arvore
arvore = clf_wine.fit(x, y)
#Avaliando a arvore 
arvore.score(x,y)
```

```python
#DÚVIDA: Como faço para mostrar todos os números de features no eixo X? (por exemplo...
#..., nesse gráfico não mostra o 1, 3, 5, 7, 9 e 11, sendo que o 9 tem um valor relevante)

#Printar a importância dos conteúdos dentro da árvore
importancia = arvore.feature_importances_
for i,v in enumerate(importancia):
	print('Feature: %d, Score: %.3f' % (i,v))

plt.bar([x for x in range(len(importancia))], importancia)
plt.show()
```

```python
#Representando a árvore por texto
text_representation = tree.export_text(clf_wine)
print(text_representation)
```

```python
#DÚVIDA: É possível mostrar a classe como o nome de um vinho ou o nome da... 
#... classe em si, ao invés de 'class_0' ou 'class_1'? Ou isso é algo já do dataset?

#Renderizando e exibindo a árvore gerada
fig = plt.figure(figsize=(50,15))
_ = tree.plot_tree(clf_wine, 
                   feature_names=wine.feature_names,  
                   class_names=wine.target_names,
                   filled=True)
```

```python
#Resultado do KNN com n = 1: ~1.0
#Resultado do KNN com n = 3: ~0.8707865168539326 <- Usaremos esse
#Resultado do KNN com n = 5: ~0.7865168539325843
#Resultado do KNN com n = 7: ~0.7471910112359551
#Resultado do KNN com n = 13: ~0.7584269662921348
#Resultado do KNN com n = 21: ~0.7247191011235955

#Criando o KNN com n = 3
neigh_wine = KNeighborsClassifier(n_neighbors=3)
neigh_wine.fit(x, y)
neigh_wine.score(x, y)
```

```python
#Dividindo a base de dados em 70% para treino e 30% para teste
X_train_wine, X_test_wine, y_train_wine, y_test_wine = train_test_split(x, y, test_size=0.3, random_state=42)

#Criando a arvore de decisão
clf_wine = DecisionTreeClassifier(random_state=1234)
#Treinando a arvore com os dados de treino
arvore = clf_wine.fit(X_train_wine, y_train_wine)
#Avaliando a arvore nos dados de teste
arvore.score(X_test_wine, y_test_wine)
```

```python
#DÚVIDA: Como faço para mostrar todos os números de features no eixo X? (por exemplo...
#..., nesse gráfico não mostra o 1, 3, 5, 7, 9 e 11, sendo que o 9 tem um valor relevante)

#Printar a importância dos conteúdos dentro da árvore
importancia = arvore.feature_importances_
for i,v in enumerate(importancia):
	print('Feature: %d, Score: %.5f' % (i,v))
 
plt.bar([x for x in range(len(importancia))], importancia)
plt.show()
```

```python
#Representando a árvore por texto
text_representation = tree.export_text(arvore)
print(text_representation)
```

```python
#Renderizando e exibindo a árvore gerada
fig = plt.figure(figsize=(50,15))
_ = tree.plot_tree(arvore,
                   filled=True)
```

```python
#Resultado do KNN com K = 1: ~0.7962962962962963
#Resultado do KNN com K = 3: ~0.7407407407407407
#Resultado do KNN com K = 5: ~0.7407407407407407
#Resultado do KNN com K = 7: ~0.7592592592592593
#Resultado do KNN com K = 13: ~0.7592592592592593
#Resultado do KNN com K = 21: ~0.7777777777777778 <- Usaremos esse
#Resultado do KNN com K = 99: ~0.7037037037037037
#Resultado do KNN com K = 123: ~0.3888888888888889

#Cria o KNN com K = 21
neigh = KNeighborsClassifier(n_neighbors=21)
#Treina o KNN na base de dados de treinamento
neigh.fit(X_train_wine, y_train_wine)
#Avalia o KNN na base de teste
neigh.score(X_test_wine, y_test_wine)
```

```python
#Criando o algoritimo de florestas aleatorias para o dataset do vinho
floresta = RandomForestClassifier()
#Treinando o algoritmo
floresta.fit(X_train_wine, y_train_wine)
#Avaliando o algoritmo
floresta.score(X_test_wine, y_test_wine)
```

```python
#Printar a importância dos conteúdos dentro da árvore (usando o algoritmo de florestas aleatorias)
importancia = floresta.feature_importances_
for i,v in enumerate(importancia):
	print('Feature: %0d, Score: %.5f' % (i,v))

plt.bar([x for x in range(len(importancia))], importancia)
plt.show()
```
